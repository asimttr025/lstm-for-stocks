```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd

#yahoo finance as data source
#pip install yfinance
import yfinance as yf

#!pip install TA-Lib
import talib as ta

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
import seaborn as sns # Visualization
sns.set_style('white', { 'axes.spines.right': False, 'axes.spines.top': False})


from keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint, TensorBoard

import math
from numpy import array

import tensorflow as tf
from datetime import datetime
from google.colab import files
```

# Import Data

```python
end_date =  '2022-05-05'
start_date = '2002-01-01'
stockname = 'GARANTI BANK'

data = yf.download('GARAN.IS', 
                      start=start_date, 
                      end= end_date, 
                      progress=False,)
```

# Adding indicators as features to our data

```python
data['RSI'] = ta.RSI(data['Close'], 14) #Adding RSI (Relative Strength Index)
data['MA'] = ta.MA(data['Close'], timeperiod = 50) #Adding 50days MA (Simple Moving Average)
data['EMA'] = ta.EMA(data['Close'],timeperiod = 50) #Adding 50days EMA (Exponential Moving Average)
data['OBV'] = ta.OBV(data['Close'],data['Volume']) #Adding OBV (On Balance Volume)
```

```python
#To get rid of null values, we start the data 50 rows ahead. (Because some indicators look fifty days back to get average)
data = data[50: len(data):]
print(len(data))

data.info()
```

# Visualizing the Closing Prices of dataframe

```python
plt.figure(figsize=(18,9))
plt.title('SchlussKursverlauf')
plt.plot(data['Close'])
plt.xlabel('Tagen (Vom StartPunkt des Datensatzes)',fontsize = 15)
plt.ylabel('SchlussKurs USD',fontsize = 15)
plt.show()
```

```python
# Extract dates (will be used in visualization)
datelist_train = data.index.get_level_values('Date')

# Indexing Batches
train_df = data.sort_values(by=['Date']).copy()

# List of considered Features
FEATURES = ['Close', 'RSI', 'OBV', 'MA', 'EMA' ]

#Create the dataset with features and filter the data to the list of FEATURES
#Specifying features
datas = pd.DataFrame(train_df)
data_filtered = datas[FEATURES]

# We add a prediction column and set dummy values to prepare the data for scaling
data_filtered_ext = data_filtered.copy()
data_filtered_ext['Prediction'] = data_filtered_ext['Close']

print('Shape of dataset == {}.'.format(data_filtered_ext.shape))
# Print the tail of the dataframe
data_filtered_ext.tail()

#Code adapted from Multivariate time series prediction example:
#https://www.relataly.com/stock-market-prediction-using-multivariate-time-series-in-python/1815/
```

# Scaling the Data

```python
from sklearn.preprocessing import StandardScaler
# Get the number of rows in the data
nrows = data_filtered.shape[0]

# Convert the data to numpy values
np_data_unscaled = np.array(data_filtered)

# Transform the data by scaling each feature to a range between 0 and 1
scaler = StandardScaler()
np_data_scaled = scaler.fit_transform(np_data_unscaled)
print(np_data_scaled[0])

# Creating a separate scaler that works on a single column for scaling predictions
scaler_pred = StandardScaler()
df_Close = pd.DataFrame(data_filtered_ext['Close'])
np_Close_scaled = scaler_pred.fit_transform(df_Close)
print(np_Close_scaled[0])
```

# Creating scaled training and test dataset





```python
#Timestep for to look n days back for predictions
timestep1 = 20

# Target Index
index_Close = data_filtered.columns.get_loc("Close")
print(index_Close)

# Determining the percentage of training data and validation data
train_data_len = math.ceil(np_data_scaled.shape[0] * 0.8)
validation_ratio = math.ceil((np_data_scaled.shape[0] - train_data_len) / 5)
validation_data_len = train_data_len + validation_ratio

# Create the training and test data
train_data1 = np_data_scaled[0:train_data_len, :]
validation_data1 = np_data_scaled[train_data_len - timestep1:validation_data_len, :]
test_data1 = np_data_scaled[validation_data_len - timestep1:, :]



print(train_data1.shape)
print(validation_data1.shape)
print(test_data1.shape)
print("")

#Function for creating time-series data
def sequencing_dataset(timestep, datax):
    x, y = [], []
    data_len = datax.shape[0]
    for i in range(timestep, data_len):
        x.append(datax[i-timestep:i,:]) #contains sequence_length values 0-sequence_length * columsn
        y.append(datax[i, index_Close]) #contains the prediction values for validation,  for single-step prediction
    
    # Convert the x and y to numpy arrays
    x = np.array(x)
    y = np.array(y)
    return x, y

# Generate training validation and test data
x_train1, y_train1 = sequencing_dataset(timestep1, train_data1)
x_validation1, y_validation1 = sequencing_dataset(timestep1, validation_data1)
x_test1, y_test1 = sequencing_dataset(timestep1, test_data1)

# Printing the shapes. Expecting result = (rows, timestep, features) (prediction value,)
print(x_train1.shape, y_train1.shape)
print(x_validation1.shape, y_validation1.shape)
print(x_test1.shape, y_test1.shape)
print("")

# The last close price of the second input sample should equal the first prediction value
print(x_train1[2][timestep1-1][index_Close])
print(y_train1[1])
print(" ")
print(x_test1[2][timestep1-1][index_Close])
print(y_test1[1])
print("")
print(x_validation1[2][timestep1-1][index_Close])
print(y_validation1[1])


#Code adapted from Multivariate time series prediction example:
#https://www.relataly.com/stock-market-prediction-using-multivariate-time-series-in-python/1815/
```

# Building The Model

**Notes:**

**EarlyStopping**: Stop training when a monitored metric has stopped
improving.

**monitor**: quantity to be monitored.

**min_delta**: minimum change in the monitored quantity to qualify as an improvement, i.e. an absolute change of less than min_delta, will count as no improvement.

**patience**: number of epochs with no improvement after which training will be stopped.

**ReduceLROnPlateau**: Reduce learning rate when a metric has stopped improving.

factor: factor by which the learning rate will be reduced. 
new_lr = lr * factor.

```python
# Configure the neural network model
model1 = Sequential()

# Model with n_neurons = inputshape Timestamps, each with x_train.shape[2] variables
n_neurons = x_train1.shape[1] * x_train1.shape[2]

print(n_neurons, x_train1.shape[1], x_train1.shape[2])
model1.add(LSTM(n_neurons,return_sequences=True, input_shape=(x_train1.shape[1], x_train1.shape[2])))
model1.add(Dropout(0.2)) 
model1.add(LSTM(n_neurons,return_sequences=False))
model1.add(Dense(1))

# Compile the model
model1.compile(optimizer='adam', loss='mse')

#CallBacks
es = EarlyStopping(monitor='val_loss',verbose = 1, patience=10)
mcp = ModelCheckpoint(filepath='weights1.h5', verbose=1, save_best_only=True, save_weights_only=True)
tb = TensorBoard('logs1')

# Training the model
epochs1 = 100
batch_size1 = 32

history1 = model1.fit(x_train1, y_train1, 
                    
                    batch_size=batch_size1, 
                    epochs=epochs1,
                    validation_data=(x_validation1, y_validation1),
                    callbacks=[es,mcp,tb])
                   
                    
                    #callbacks=[early_stop])

# Plot training & validation loss values
fig, ax = plt.subplots(figsize=(16, 5), sharex=True)
sns.lineplot(data=history1.history["loss"])
#validation loss show
plt.title("Model loss")
plt.ylabel("Loss")
plt.xlabel("Epoch")
ax.xaxis.set_major_locator(plt.MaxNLocator(epochs1))
plt.legend(["Train", "Validation"], loc="upper left")
plt.grid()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error # Packages for measuring model performance / errors
# Get the predicted values
y_pred_scaled1 = model1.predict(x_test1)
y_pred_scaled2 = model2.predict(x_test2)
y_pred_scaled3 = model3.predict(x_test3)



print(x_test1.shape)
print(y_pred_scaled1.shape)
print("")
print(x_test2.shape)
print(y_pred_scaled2.shape)
print("")
print(x_test3.shape)
print(y_pred_scaled3.shape)

print("--------------------------------------------------------")

# Unscale the predicted values
y_pred1 = scaler_pred.inverse_transform(y_pred_scaled1)
y_pred2 = scaler_pred.inverse_transform(y_pred_scaled2)
y_pred3 = scaler_pred.inverse_transform(y_pred_scaled3)



y_test_unscaled1 = scaler_pred.inverse_transform(y_test1.reshape(-1, 1))
y_test_unscaled2 = scaler_pred.inverse_transform(y_test2.reshape(-1, 1))
y_test_unscaled3 = scaler_pred.inverse_transform(y_test3.reshape(-1, 1))


# Mean Absolute Error (MAE)
MAE1 = mean_absolute_error(y_test_unscaled1, y_pred1)
MAE2 = mean_absolute_error(y_test_unscaled2, y_pred2)
MAE3 = mean_absolute_error(y_test_unscaled3, y_pred3)


print(f'Median Absolute Error for Model1 with timestep 20 (MAE): {np.round(MAE1, 4)}')
print(f'Median Absolute Error for Model2 with timestep 50 (MAE): {np.round(MAE2, 4)}')
print(f'Median Absolute Error for Model2 with timestep 100 (MAE): {np.round(MAE3, 4)}')


print("--------------------------------------------------------")

#Root Mean Squared Error (RMSE)
RMSE1 = math.sqrt(mean_squared_error(y_test_unscaled1, y_pred1))
RMSE2 = math.sqrt(mean_squared_error(y_test_unscaled2, y_pred2))
RMSE3 = math.sqrt(mean_squared_error(y_test_unscaled3, y_pred3))


print(f'Root Mean Squared Error (RMSE) for Model1 with timestep 20 : {np.round(RMSE1, 4)}')
print(f'Root Mean Squared Error (RMSE) for Model2 with timestep 50 : {np.round(RMSE2, 4)}')
print(f'Root Mean Squared Error (RMSE) for Model2 with timestep 100 : {np.round(RMSE3, 4)}')

print("--------------------------------------------------------")

# Mean Absolute Percentage Error (MAPE)
MAPE1 = np.mean((np.abs(np.subtract(y_test_unscaled1, y_pred1)/ y_test_unscaled1))) * 100
MAPE2 = np.mean((np.abs(np.subtract(y_test_unscaled2, y_pred2)/ y_test_unscaled2))) * 100
MAPE3 = np.mean((np.abs(np.subtract(y_test_unscaled3, y_pred3)/ y_test_unscaled3))) * 100

print(f'Mean Absolute Percentage Error for Model1 with timestep 20  (MAPE): {np.round(MAPE1, 4)} %')
print(f'Mean Absolute Percentage Error for Model2 with timestep 50  (MAPE): {np.round(MAPE2, 4)} %')
print(f'Mean Absolute Percentage Error for Model2 with timestep 100  (MAPE): {np.round(MAPE3, 4)} %')


print("--------------------------------------------------------")

# Median Absolute Percentage Error (MDAPE)
MDAPE1 = np.median((np.abs(np.subtract(y_test_unscaled1, y_pred1)/ y_test_unscaled1)) ) * 100
MDAPE2 = np.median((np.abs(np.subtract(y_test_unscaled2, y_pred2)/ y_test_unscaled2)) ) * 100
MDAPE3 = np.median((np.abs(np.subtract(y_test_unscaled3, y_pred3)/ y_test_unscaled3)) ) * 100


print(f'Median Absolute Percentage Error for Model1 with timestep 20  (MDAPE): {np.round(MDAPE1, 4)} %')
print(f'Median Absolute Percentage Error for Model2 with timestep 50  (MDAPE): {np.round(MDAPE2, 4)} %')
print(f'Median Absolute Percentage Error for Model2 with timestep 100  (MDAPE): {np.round(MDAPE3, 4)} %')

#Code adapted from Multivariate time series prediction example:
#https://www.relataly.com/stock-market-prediction-using-multivariate-time-series-in-python/1815/
```
